```python
# 컬럼 함수를 이용시 타입은 컬럼으로
c = col('강도 검거')
print(type(c))
```

```python
# when (조건, 값)
df.withColumn('정도',
            when(df['살인 발생']>=10, '많음').\
            when(df['살인 발생']>=5, '보통').\
            otherwise('적음')  
             ).toPandas()
```

### 함수 적용 udf

```python
ft = udf(lambda v : '많음' if v >=10 else '적음')
df.withColumn('정도1',ft(df['살인 발생'])).toPandas()
```

```python
# regexp_replace : 특정 패턴을 원하는 패턴으로 변경 가능
# 타입은 변하지 않음
df = df.withColumn('폭력 발생', regexp_replace('폭력 발생',',',''))
df.toPandas()
```

```python
# 타입 변경
df = df.withColumn('절도 발생',df['절도 발생'].cast(IntegerType()))
df.printSchema()
```
